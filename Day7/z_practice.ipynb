### ❓ Q1. Vectorized Operations
Given: `prices = np.array([100, 250, 75, 300, 150, 220])`

- (a) Apply a **10% discount** on all prices.
- (b) Add a **flat ₹20 shipping charge** to all prices.
- (c) Find prices after **both** applying 10% discount and then adding ₹20 shipping.

````python
import numpy as np

# Given array
prices = np.array([100, 250, 75, 300, 150, 220])

discounted_prices = prices * 0.90
print("After 10% discount:")
print(discounted_prices)

shipping_prices = prices + 20
print("\nAfter adding ₹20 shipping:")
print(shipping_prices)

final_prices = (prices * 0.90) + 20
print("\nAfter 10% discount and ₹20 shipping:")
print(final_prices)
````

### ❓ Q2. Aggregation Functions
Given test scores of 8 students: `[62, 78, 55, 90, 83, 47, 71, 95]`

Calculate and print:
- (a) Total marks (sum)
- (b) Class average (mean)
- (c) Highest and lowest marks
- (d) Median score
- (e) Standard deviation (how spread out are the marks?)

````python
import numpy as np

scores = np.array([62, 78, 55, 90, 83, 47, 71, 95])

total = np.sum(scores)
print("Total marks:", total)

average = np.mean(scores)
print("Class average:", average)

highest = np.max(scores)
lowest = np.min(scores)
print("Highest marks:", highest)
print("Lowest marks:", lowest)

median = np.median(scores)
print("Median score:", median)

std_dev = np.std(scores)
print("Standard deviation:", std_dev)
````

### ❓ Q3. Aggregation on 2D Array
Create this marks table (3 students, 3 subjects):
```
[[80, 70, 90],
 [60, 85, 75],
 [95, 65, 88]]
```
- (a) Find the **total score per student** (sum across each row).
- (b) Find the **average score per subject** (mean down each column).
- (c) Find the **overall maximum** score in the entire table.

````python
import numpy as np

marks = np.array([
    [80, 70, 90],
    [60, 85, 75],
    [95, 65, 88]
])

student_totals = np.sum(marks, axis=1)
print("Total score per student:", student_totals)

subject_averages = np.mean(marks, axis=0)
print("Average score per subject:", subject_averages)

max_score = np.max(marks)
print("Overall maximum score:", max_score)
````

### ❓ Q4. Boolean Masking
Given temperatures: `temps = np.array([22, 35, 18, 40, 28, 15, 38, 25, 42, 30])`

- (a) Find all temperatures **above 30°C** (hot days).
- (b) Find all temperatures **between 20°C and 30°C** (comfortable days).
- (c) Count how many days were **below 20°C** (cold days).
- (d) Check if **any** day had temperature **above 40°C**.

````python
import numpy as np

temps = np.array([22, 35, 18, 40, 28, 15, 38, 25, 42, 30])

hot_days = temps[temps > 30]
print("Hot days (above 30°C):", hot_days)

comfortable_days = temps[(temps >= 20) & (temps <= 30)]
print("Comfortable days (20°C to 30°C):", comfortable_days)

cold_days_count = np.sum(temps < 20)
print("Number of cold days (below 20°C):", cold_days_count)

above_40 = np.any(temps > 40)
print("Any day above 40°C?", above_40)
````

### ❓ Q5. `np.where()` — Grade Assignment
Given scores: `scores = np.array([45, 62, 78, 91, 55, 83, 70, 38, 95, 66])`

Use `np.where()` to:
- (a) Label each score as `"Pass"` (≥ 60) or `"Fail"` (< 60).
- (b) Apply a **curved score**: Add 5 marks to anyone who scored **below 50**, leave others unchanged.
- (c) Assign grade **"Distinction"** if score ≥ 85, **"Merit"** if score ≥ 65, else **"Pass/Fail"** as appropriate.

````python
import numpy as np

scores = np.array([45, 62, 78, 91, 55, 83, 70, 38, 95, 66])

result = np.where(scores >= 60, "Pass", "Fail")
print("Pass/Fail:", result)

curved_scores = np.where(scores < 50, scores + 5, scores)
print("Curved scores:", curved_scores)

grades = np.where(
    scores >= 85, "Distinction",
    np.where(scores >= 65, "Merit", np.where(scores >= 60, "Pass", "Fail"))
)

print("Grades:", grades)
````

### ❓ 6. Mini Challenge — Put It All Together! 🏆

You are a teacher with the following data for **6 students** across **4 subjects**:  
*(Subjects: Maths, Science, English, Computer)*

```
Student 0: [82, 76, 90, 88]
Student 1: [55, 60, 72, 50]
Student 2: [91, 85, 78, 95]

```

Do the following:
- (a) Print the shape of the array.
- (b) Find the **total marks** for each student.
- (c) Find the **average marks** per subject.
- (d) Identify students whose **total marks are below 250** (struggling students).
- (e) Use `np.where()` to label each student's **total** as `"Good"` if ≥ 300, else `"Needs Improvement"`.

````python
import numpy as np

marks = np.array([
    [82, 76, 90, 88],
    [55, 60, 72, 50],
    [91, 85, 78, 95]
])
print("Shape of the array:", marks.shape)

total_marks = np.sum(marks, axis=1)
print("Total marks for each student:", total_marks)

subject_average = np.mean(marks, axis=0)
print("Average marks per subject:", subject_average)

struggling_students = np.where(total_marks < 250)[0]
print("Students with total marks below 250:", struggling_students)

labels = np.where(total_marks >= 300, "Good", "Needs Improvement")
print("Student Labels:", labels)
````
