### ❓ Q1. reshape() and flatten()
- (a) Create a 1D array of numbers from **1 to 24** using `np.arange()`.
- (b) Reshape it into a **4×6** matrix.
- (c) Flatten it back to 1D using `flatten()`.
- (d) Reshape the same array into shape **(2, 3, 4)** — a 3D array.

```python
import numpy as np

arr = np.arange(1, 25)
print("1D Array:")
print(arr)

matrix = arr.reshape(4, 6)
print("\n4x6 Matrix:")
print(matrix)

flat = matrix.flatten()
print("\nFlattened Array:")
print(flat)

array3d = arr.reshape(2, 3, 4)
print("\n3D Array (2x3x4):")
print(array3d)
```

### ❓ Q4. Percentiles and IQR
Given salaries (in thousands): `[22, 28, 35, 40, 42, 45, 50, 55, 60, 250]`
- (a) Compute **Q1, Q2 (Median), and Q3**.
- (b) Compute the **IQR**.
- (c) Find and print any **outliers** using the IQR method.
- (d) Compute the **90th percentile** salary.

```python
import numpy as np

salaries = np.array([22, 28, 35, 40, 42, 45, 50, 55, 60, 250])

q1 = np.percentile(salaries, 25)
q2 = np.percentile(salaries, 50)
q3 = np.percentile(salaries, 75)

print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)

iqr = q3 - q1
print("IQR:", iqr)

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

outliers = salaries[(salaries < lower) | (salaries > upper)]
print("Outliers:", outliers)

p90 = np.percentile(salaries, 90)
print("90th Percentile:", p90)
```

### ❓ Q5. Missing Values
Given temperatures: `[28.5, np.nan, 31.0, np.nan, 29.5, 27.0, np.nan, 33.0, 30.5, 26.0]`
- (a) Count how many values are missing.
- (b) Compute the **mean, max, and min** ignoring NaN.
- (c) Replace all NaN values with the **mean temperature**.
- (d) Verify no NaN values remain.

```python
import numpy as np

temps = np.array([28.5, np.nan, 31.0, np.nan, 29.5,
                  27.0, np.nan, 33.0, 30.5, 26.0])

missing = np.isnan(temps).sum()
print("Missing values:", missing)

mean_temp = np.nanmean(temps)
max_temp = np.nanmax(temps)
min_temp = np.nanmin(temps)

print("Mean temperature:", mean_temp)
print("Maximum temperature:", max_temp)
print("Minimum temperature:", min_temp)

temps = np.where(np.isnan(temps), mean_temp, temps)
print("\nArray after replacing NaN:")
print(temps)
print("\nAny NaN values remaining?", np.isnan(temps).any())
```

### ❓ Q6. Sorting
Given: `scores = np.array([88, 55, 92, 70, 45, 78, 61, 85])`  
        `names  = np.array(['Alice','Bob','Charlie','Diana','Eve','Frank','Grace','Hank'])`
- (a) Sort the scores in **ascending** order.
- (b) Sort the scores in **descending** order.
- (c) Use `np.argsort()` to print names with their scores, **sorted from highest to lowest**.
- (d) Print the **top 3 students** by score.

```python
import numpy as np

scores = np.array([88, 55, 92, 70, 45, 78, 61, 85])
names = np.array(['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Hank'])

ascending = np.sort(scores)
print("Ascending order:", ascending)

descending = np.sort(scores)[::-1]
print("Descending order:", descending)

indices = np.argsort(scores)[::-1]

print("\nStudents sorted by score (Highest to Lowest):")
for i in indices:
    print(names[i], ":", scores[i])

print("\nTop 3 Students:")
for i in indices[:3]:
    print(names[i], ":", scores[i])
```

### ❓ Q7. Unique Values
Given city data: `cities = np.array(['Delhi','Mumbai','Delhi','Chennai','Bangalore','Mumbai','Delhi','Kolkata','Chennai','Mumbai'])`
- (a) Find all **unique cities**.
- (b) Find how many times **each city appears**.
- (c) Find which city appears the **most**.
- (d) Count **total unique cities**.

```python
import numpy as np

cities = np.array([
    'Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Bangalore',
    'Mumbai', 'Delhi', 'Kolkata', 'Chennai', 'Mumbai'
])

unique_cities = np.unique(cities)
print("Unique cities:", unique_cities)

unique, counts = np.unique(cities, return_counts=True)
print("\nCity counts:")
for city, count in zip(unique, counts):
    print(city, ":", count)

most_city = unique[np.argmax(counts)]
most_count = np.max(counts)
print("\nMost frequent city:", most_city)
print("Appears:", most_count, "times")

total_unique = len(unique)
print("\nTotal unique cities:", total_unique)
```
