## 📝 Exercise 1: Student Information Form
### Instructions:
1. Ask the user for their **Name**, **Age**, and **City** using `input()`.
2. Convert the **Age** to an integer.
3. Print a summary message using an **f-string**.

*Expected Output Format:*
```text
My name is Rahul
My age is 25
My city is Delhi
```


````python
# Exercise 1: Student Information Form
# Write your code here (ask for name, age, city and print using f-string):

Name = input("Enter your name: ")
Age = input("Enter your age: ")
City = input("Enter your city: ")
Age = int(Age)
print(f"My name is {Name}")
print(f"My age is {Age}")
print(f"My city is {City}")
````

## 📝 Exercise 2: Simple Calculator
### Instructions:
1. Ask the user for two numbers using `input()`.
2. Convert them to integers.
3. Calculate and print the results for **Addition**, **Subtraction**, **Multiplication**, and **Division**.


````python
# Exercise 2: Simple Calculator
# Write your code here (ask for two numbers, convert, and print calculations):

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

sum_result = num1 + num2
diff_result = num1 - num2
multi_result = num1 * num2
quot_result = num1 / num2

print(f"Addition: {sum_result}")
print(f"Subtraction: {diff_result}")
print(f"Multiplication: {multi_result}")
print(f"Division: {quot_result}")
````

## 📝 Exercise 3: String Practice
### Instructions:
1. Ask the user to enter their full name.
2. Print the **length** of their name.
3. Print the name in **UPPERCASE**.
4. Print the name in **lowercase**.


````python
# Exercise 3: String Practice
# Write your code here (ask for name and perform string operations):

name = input("Enter your full name: ")
print(f"Length of name: {len(name)}")
print(f"Uppercase: {name.upper()}")
print(f"Lowercase: {name.lower()}")
````

## 📝 Exercise 4: List Practice
### Instructions:
1. Create a list called `fruits` containing: `"Apple"`, `"Mango"`, `"Banana"`.
2. Append `"Orange"` to the list.
3. Insert `"Kiwi"` at index `1`.
4. Remove `"Banana"` from the list.
5. Sort the list alphabetically.
6. Print the final list.


````python
# Exercise 4: List Practice
# Write your code here (create list and modify using methods):

my_list = ["Apple","Mango","Banana"]
print("Original List: ", my_list)
my_list.append("Orange")
my_list[1] = "Kiwi"
my_list.remove("Banana")
print("Modified List: ", my_list)
````

## 📝 Exercise 5: Dictionary Practice
### Instructions:
1. Create a student dictionary with keys `'name'` (value `'Rahul'`) and `'age'` (value `22`).
2. Add a new key `'city'` with the value `'Delhi'`.
3. Print the keys of the dictionary.
4. Print the values of the dictionary.
5. Print the value of the `'name'` key using the `.get()` method.


````python
# Exercise 5: Dictionary Practice
# Write your code here (create dict, add key, print keys/values and get name):
# Create a student dictionary
student = {
    "name": "Rahul",
    "age": 22
}
student["city"] = "Delhi"
print(student.keys())
print(student.values())
print("Name:", student.get("name"))
````
