## 📝 Exercise 1: Student Information Form
### Instructions:
1. Ask the user for their **Name**, **Age**, and **City** using `input()`.
2. Convert the **Age** to an integer.
3. Print a summary message using an **f-string**.

*Expected Output Format:*
```text
My name is Rahul
My age is 25
My city is Delhi
```


````python
# Exercise 1: Student Information Form
# Write your code here (ask for name, age, city and print using f-string):

Name = input("Enter your name: ")
Age = input("Enter your age: ")
City = input("Enter your city: ")
Age = int(Age)
print(f"My name is {Name}")
print(f"My age is {Age}")
print(f"My city is {City}")
````

## 📝 Exercise 2: Simple Calculator
### Instructions:
1. Ask the user for two numbers using `input()`.
2. Convert them to integers.
3. Calculate and print the results for **Addition**, **Subtraction**, **Multiplication**, and **Division**.


````python
# Exercise 2: Simple Calculator
# Write your code here (ask for two numbers, convert, and print calculations):

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

sum_result = num1 + num2
diff_result = num1 - num2
multi_result = num1 * num2
quot_result = num1 / num2

print(f"Addition: {sum_result}")
print(f"Subtraction: {diff_result}")
print(f"Multiplication: {multi_result}")
print(f"Division: {quot_result}")
````

## 📝 Exercise 5: Dictionary Practice
### Instructions:
1. Create a student dictionary with keys `'name'` (value `'Rahul'`) and `'age'` (value `22`).
2. Add a new key `'city'` with the value `'Delhi'`.
3. Print the keys of the dictionary.
4. Print the values of the dictionary.
5. Print the value of the `'name'` key using the `.get()` method.


````python
# Exercise 5: Dictionary Practice
# Write your code here (create dict, add key, print keys/values and get name):
# Create a student dictionary
student = {
    "name": "Rahul",
    "age": 22
}
student["city"] = "Delhi"
print(student.keys())
print(student.values())
print("Name:", student.get("name"))
````


## 📝 Exercise 6: Strings (`str`)

- **Q1.1: Name Formatter**
  Write a code snippet that takes a messy name input (e.g., `name = "   praneet bawa   "`), trims the leading/trailing spaces, and capitalizes the first letter of each word (Title Case).
  *Expected output:* `"Praneet Bawa"`

- **Q1.2: Vowel Replacer**
  Given the string `sentence = "python is fun"`, replace all occurrences of the letter `"o"` with an asterisk `"*"` and print the result.
  *Expected output:* `"pyth*n is fun"`

- **Q1.3: Check Substring**
  Given `text = "Data Science Plan"`, check if the word `"Science"` exists in the string. Print `True` if it is present, and `False` otherwise.



````python
# Name Formatter

name = "   praneet bawa   "

name = name.strip()
name = name.title()

print("Name:", name)
````

````python
# Vowel Replacer

sentence = "python is fun"

sentence = sentence.replace("o", "*")

print("Modified sentence:", sentence)
````

````python
# Check Substring

text = "Data Science Plan"

print("Science" in text)
````

## 📝 Exercise 6: Lists (`list`)

- **Q2.1: Color Swapper**
  Start with the list `colors = ["red", "green", "blue"]`.
  1. Replace the middle element (`"green"`) with `"yellow"`.
  2. Add `"purple"` to the end of the list.
  3. Print the updated list.
  *Expected output:* `['red', 'yellow', 'blue', 'purple']`

- **Q2.2: List Math**
  Given the list `numbers = [12, 45, 2, 89, 30]`, write code to find and print:
  1. The total count of elements (length).
  2. The maximum value in the list.
  3. The sum of all elements.

- **Q2.3: Slice and Reverse**
  Given `items = ["pen", "pencil", "eraser", "ruler", "marker"]`, extract a slice containing `["pencil", "eraser", "ruler"]` and reverse its elements *without* modifying the original list.

````python
# Color Swapper

colors = ["red", "green", "blue"]

colors[1] = "yellow"

colors.append("purple")

print("Updated colors:", colors)
````

````python
# List Math

numbers = [12, 45, 2, 89, 30]

print("Length of numbers:", len(numbers))

print("Maximum value:", max(numbers))

print("Sum of numbers:", sum(numbers))
````

````python
# Slice and Reverse

items = ["pen", "pencil", "eraser", "ruler", "marker"]

new_list = items[1:4]
new_list.reverse()

print("Reversed slice: ",new_list)

print("Original items: ", items)
````
