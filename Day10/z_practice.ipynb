### ❓ Q8. NumPy with Pandas
- (a) Create a NumPy array of shape **(4, 3)** with random integers between 60 and 100 (seed=7).
- (b) Convert it to a **Pandas DataFrame** with columns `['Physics', 'Chemistry', 'Biology']` and row index `['S1','S2','S3','S4']`.
- (c) Convert the `'Chemistry'` column back to a **NumPy array**.
- (d) Apply `np.mean()`, `np.max()`, and `np.min()` on that NumPy array.

````python
import numpy as np
import pandas as pd

np.random.seed(7)

arr = np.random.randint(60, 101, size=(4, 3))
print("NumPy Array:")
print(arr)

df = pd.DataFrame(
    arr,
    columns=['Physics', 'Chemistry', 'Biology'],
    index=['S1', 'S2', 'S3', 'S4']
)

print("\nDataFrame:")
print(df)

chemistry_array = df['Chemistry'].to_numpy()

print("\nChemistry NumPy Array:")
print(chemistry_array)

print("\nMean:", np.mean(chemistry_array))
print("Max:", np.max(chemistry_array))
print("Min:", np.min(chemistry_array))
````

### ❓ Q9. Combined Operations
Create **50 random test scores** between 30 and 100 using `np.random.randint()` with seed=99.
- (a) Print the **mean, median, std deviation**.
- (b) Find the **25th and 75th percentile**.
- (c) Find the **top 10 scores** (hint: sort descending, take first 10).
- (d) Replace all scores **below 40** with `np.nan` (simulate absent students).
- (e) Recompute the mean using the appropriate NaN-safe function.

````python
import numpy as np

np.random.seed(99)

scores = np.random.randint(30, 101, 50)

print("Scores:")
print(scores)

print("\nMean:", np.mean(scores))
print("Median:", np.median(scores))
print("Standard Deviation:", np.std(scores))

print("\n25th Percentile:", np.percentile(scores, 25))
print("75th Percentile:", np.percentile(scores, 75))

top_10 = np.sort(scores)[::-1][:10]
print("\nTop 10 Scores:")
print(top_10)

scores_nan = scores.astype(float)
scores_nan[scores_nan < 40] = np.nan

print("\nScores after replacing below 40 with NaN:")
print(scores_nan)

print("\nMean after removing NaN:", np.nanmean(scores_nan))
````

### ❓ Q10. Mini Challenge — Sales Report 🏆

A company has **monthly sales data** (in ₹ lakhs) for **4 products** over **6 months**:

```
           Jan   Feb   Mar   Apr   May   Jun
Product A:  12    15   np.nan  18    20    22
Product B:  8     10    11   np.nan  14    16
Product C:  20    18    22    25    23    np.nan
Product D:  5    np.nan  7     9     8     10
```

- (a) Count total **missing values** in the dataset.
- (b) Find the **total sales per product** (ignore missing months).
- (c) Find the **best performing product** (highest total).
- (d) Find the **average sales per month** across all products.
- (e) Reshape the sales data into **shape (2, 12)** (after filling NaN with 0).
- (f) Use `np.where()` to label each product as `"Top Seller"` if total > 90, else `"Regular"`.

````python
import numpy as np
import pandas as pd

sales = np.array([
    [12, 15, np.nan, 18, 20, 22],
    [8, 10, 11, np.nan, 14, 16],
    [20, 18, 22, 25, 23, np.nan],
    [5, np.nan, 7, 9, 8, 10]
])

products = np.array(['Product A', 'Product B', 'Product C', 'Product D'])

df = pd.DataFrame(
    sales,
    index=products,
    columns=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
)

print("Sales Data:")
print(df)

print("\nMissing Values:", np.isnan(sales).sum())

totals = np.nansum(sales, axis=1)
print("\nTotal Sales per Product:")
for product, total in zip(products, totals):
    print(product, ":", total)

best_product = products[np.argmax(totals)]
print("\nBest Performing Product:", best_product)

monthly_average = np.nanmean(sales, axis=0)
print("\nAverage Sales per Month:")
print(monthly_average)

sales_filled = np.nan_to_num(sales)
reshaped_sales = sales_filled.reshape(2, 12)

print("\nReshaped Sales (2,12):")
print(reshaped_sales)

labels = np.where(totals > 90, "Top Seller", "Regular")

print("\nProduct Labels:")
for product, label in zip(products, labels):
    print(product, ":", label)
````
