Question 1: Student Result Analyzer (10 Marks)
Create a function that accepts a list of student marks.
1. Calculate the total marks.
2. Calculate the average marks.
3. Find the highest and lowest marks.
4. Count how many students passed (marks ‡ 40).
5. Return all the calculated values.
Example: marks = [78, 45, 32, 91, 67, 39]
Expected Output:
Total Marks: 352
Average Marks: 58.67
Highest Marks: 91
Lowest Marks: 32
Passed Students: 4

```python
def student_result_analyzer(marks):
    total = sum(marks)
    average = total / len(marks)
    highest = max(marks)
    lowest = min(marks)
    passed = sum(1 for mark in marks if mark >= 40)

    return total, average, highest, lowest, passed

marks = [78, 45, 32, 91, 67, 39]

total, average, highest, lowest, passed = student_result_analyzer(marks)

print("Total Marks:", total)
print("Average Marks:", round(average, 2))
print("Highest Marks:", highest)
print("Lowest Marks:", lowest)
print("Passed Students:", passed)
```

Question 2: Shopping Cart Calculator (10 Marks)
Create a function that accepts a list of item prices and a discount percentage (default = 10).
1. Calculate the total bill.
2. Apply the discount.
3. If the final bill is greater than n5000, apply an additional 5% discount.
4. Return the final payable amount.
Example: prices = [1200, 800, 1500, 2500]
Expected Output:
Original Bill: 6000
Discount Applied: 10%
Extra Discount: Yes
Final Amount: 5130

```python
def shopping_cart(prices, discount=10):
    total = sum(prices)

    final_amount = total - (total * discount / 100)

    extra_discount = "No"

    if final_amount > 5000:
        final_amount = final_amount - (final_amount * 5 / 100)
        extra_discount = "Yes"

    return total, discount, extra_discount, final_amount


prices = [1200, 800, 1500, 2500]

bill, disc, extra, final = shopping_cart(prices)

print("Original Bill:", bill)
print("Discount Applied:", str(disc) + "%")
print("Extra Discount:", extra)
print("Final Amount:", round(final, 2))
```

Question 3: Employee Information System (10 Marks)
Create a function using **kwargs to store employee details.
1. Print all employee details.
2. If salary is greater than n50,000, print 'High Salary', otherwise print 'Normal Salary'.
3. Count the total number of details received.
4. Return the total count.
Example:
employee(name='Rahul', age=25, department='IT', salary=60000)
Expected Output:
Name: Rahul
Age: 25
Department: IT
Salary: 60000
High Salary
Total Details: 4


```python
def employee(**kwargs):
    for key, value in kwargs.items():
        print(f"{key.capitalize()}: {value}")

    if kwargs.get("salary", 0) > 50000:
        print("High Salary")
    else:
        print("Normal Salary")

    total_details = len(kwargs)
    print("Total Details:", total_details)

    return total_details

employee(
    name="Rahul",
    age=25,
    department="IT",
    salary=60000
)
```

Question 4: Number Analysis Tool (20 Marks)
Create a function that accepts any number of integers using *args.
1. Find the largest and smallest number.
2. Count even and odd numbers.
3. Calculate the sum and average.
4. Print all numbers divisible by both 3 and 5.
5. Return all calculated values.
Example: analyze_numbers(12, 15, 20, 9, 30, 7, 4)
Expected Output:
Largest Number: 30
Smallest Number: 4
Even Numbers: 4
Odd Numbers: 3
Sum: 97
Average: 13.86
Divisible by both 3 and 5: 15, 30

```python
def analyze_numbers(*args):
    largest = max(args)
    smallest = min(args)

    even = 0
    odd = 0

    for num in args:
        if num % 2 == 0:
            even += 1
        else:
            odd += 1

    total = sum(args)
    average = total / len(args)

    divisible = []
    for num in args:
        if num % 3 == 0 and num % 5 == 0:
            divisible.append(num)

    print("Largest Number:", largest)
    print("Smallest Number:", smallest)
    print("Even Numbers:", even)
    print("Odd Numbers:", odd)
    print("Sum:", total)
    print("Average:", round(average, 2))
    print("Divisible by both 3 and 5:", *divisible, sep=", ")

    return largest, smallest, even, odd, total, average, divisible

analyze_numbers(12, 15, 20, 9, 30, 7, 4)
```
