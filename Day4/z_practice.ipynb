### Exercise 3: Filter and Sort with Lambdas
You are given a list of dictionaries representing employees:
```python
employees = [
    {"name": "Amit", "age": 28, "salary": 50000},
    {"name": "Neha", "age": 35, "salary": 75000},
    {"name": "Raj", "age": 24, "salary": 45000},
    {"name": "Pooja", "age": 31, "salary": 80000}
]
```
Write a program that uses:
1. A lambda function with `filter()` to get employees with a salary greater than `50,000`.
2. A lambda function with `sorted()` to sort the filtered employees by their `age` in ascending order.

### Exercise 4: Global Counter
Write a function `update_count()` that increments a global variable `runs_count` by 1 every time it is called, and prints the current value.


````python
# Exercise 1: Leap Year Checker Function
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False


year = int(input("Enter a year: "))

if is_leap_year(year):
    print(year, "is a Leap Year")
else:
    print(year, "is not a Leap Year")
````

````python
# Exercise 2: Statistics Calculator (Multiple Returns)
def calculate_stats(numbers):
    total = sum(numbers)
    count = len(numbers)
    average = total / count

    return total, average, count


numbers = [10, 20, 30, 40, 50]

total, average, count = calculate_stats(numbers)

print("Sum:", total)
print("Average:", average)
print("Count:", count)
````

````python
# Exercise 3: Filter and Sort with Lambdas
employees = [
    {"name": "Amit", "age": 28, "salary": 50000},
    {"name": "Neha", "age": 35, "salary": 75000},
    {"name": "Raj", "age": 24, "salary": 45000},
    {"name": "Pooja", "age": 31, "salary": 80000}
]

filtered = list(filter(lambda emp: emp["salary"] > 50000, employees))

sorted_employees = sorted(filtered, key=lambda emp: emp["age"])

print("Employees with salary greater than 50000:")

for emp in sorted_employees:
    print(emp)
````

````python
# Exercise 4: Global Counter
runs_count = 0

def update_count():
    global runs_count
    runs_count = runs_count + 1
    print("Current Count:", runs_count)


update_count()
update_count()
update_count()
````
