# Python for Beginners: Control Flow Statements

Welcome to this interactive Jupyter Notebook! This guide is designed to help you master Python's **Control Flow Statements**.
Control flow refers to the order in which individual statements, instructions, or function calls of an imperative program are executed. By default, a Python program runs line-by-line from top to bottom. However, control flow statements allow you to bypass this sequential execution, enabling decision-making and repetitive tasks.

In this notebook, we will cover:
1. **Conditional Branching**: `if`, `if-else`, `elif`, and Nested Conditions.
2. **Looping structures**: `for` loops and `while` loops.
3. **Loop Control Statements**: `break`, `continue`, and `pass`.

## Module 1: Conditional Statements
Conditional statements allow your program to make decisions based on whether a condition is `True` or `False`. They act as decision forks in your code.

We will explore four forms:
1. **Simple `if` statement**
2. **`if-else` statement**
3. **`if-elif-else` statement**
4. **Nested `if` conditions**

Let's look at each one in detail.


### 1.1 The `if` Statement
The `if` statement is the simplest form of decision-making. It runs a block of code **only if** the specified condition evaluates to `True`.

#### Syntax:
```python
if condition:
    # Code to run if condition is True
```

> **⚠️ Indentation matters!** Python uses 4 spaces (or a tab) of indentation to define the block of code inside the conditional statement. If you forget to indent, Python will throw an `IndentationError`.

#### How it works:
```text
      [Start]
         │
         ▼
  Is condition True?  ───── No ───┐
         │                        │
        Yes                       │
         │                        │
         ▼                        │
   [Run if-block]                 │
         │                        │
         ├────────────────────────┘
         ▼
    [Continue]
```

Let's see this in action:


````python
# A simple check for passing score
score = 5

if score >= 50:
    print("Congratulations!")
    print("You passed the exam.")

print("This line runs regardless of the condition.")
````

### 1.2 The `if-else` Statement
The `if-else` statement provides an alternative block of code that runs **only if** the `if` condition evaluates to `False`. It's a binary decision maker.

#### Syntax:
```python
if condition:
    # Code to run if condition is True
else:
    # Code to run if condition is False
```

#### How it works:
```text
               [Start]
                  │
                  ▼
         Is condition True?
             /        \
          Yes          No
          /            \
         ▼              ▼
   [Run if-block]   [Run else-block]
         \              /
          \            /
           ▼          ▼
            [Continue]
```

Let's run code to see this:


````python
age = 2

if age >= 18:
    print("You are eligible to vote!")
else:
    print("You are not old enough to vote yet.")
    print(f"Please wait {18 - age} more year(s).")
print("i am outside the block")
````

### 1.3 The `if-elif-else` Statement (Chained Conditions)
When you have more than two possibilities, you can chain multiple conditions using `elif` (short for **else if**). Python checks each condition from top to bottom and runs the first block that evaluates to `True`. If none of them are true, the final `else` block runs.

#### Syntax:
```python
if condition1:
    # Code for condition1
elif condition2:
    # Code for condition2
elif condition3:
    # Code for condition3
else:
    # Default code if none of the above are True
```

Let's test this with a grading system:


````python
# Grade classification
marks = 98


if marks >= 90:
    grade = "A"
elif marks >= 80:
    grade = "B"
elif marks >= 70:
    grade = "C"
elif marks >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Marks: {marks}, Grade: {grade}")
````

### 1.4 Nested Conditions
You can place one conditional statement inside another. This is called nesting. It is useful when a decision depends on multiple subsequent conditions.

#### ⚠️ Warning:
Deep nesting can make code hard to read. Where possible, try to simplify nested conditions using logical operators (`and`, `or`).

#### Example Scenario:
Let's check if a candidate is eligible for a loan:
1. They must be at least 18 years old.
2. If they are 18+, we check their monthly income:
   * If income >= 50,000, they are approved for a high limit.
   * If income is less, they are approved for a standard limit.
3. If they are under 18, they are not eligible.


````python
age = 2
income = 5000

if age >= 18:
    print("Age verified: Adult.")
    # Nested condition starts here
    if income >= 50000:
        print("Loan status: Approved for premium limit!")
    else:
        print("Loan status: Approved for standard limit.")
else:
    print("Age verified: Minor.")
    print("Loan status: Denied (Must be 18+).")
````

````python
gender="Female"
marks=50

if(gender=="Female"):
    if(marks>90):
        print("congrats!..u got 100 percent scholarship")
    elif(marks>70):
        print("congrats!..u got 50 percent scholarship")
    else:
        print("sorry! minimum marks must be 70")
else:
    print("this scholarship is only for girls")
````

## Module 2: Loops (Iteration)
Loops allow you to run a block of code repeatedly. Python provides two main types of loops:
1. **`for` loops**: Best when you know in advance how many times you want to iterate (definite iteration).
2. **`while` loops**: Best when you want to iterate until a specific condition changes (indefinite iteration).


### 2.1 The `for` Loop
A `for` loop is used to iterate over a sequence (such as a list, tuple, string, or range). It executes the loop body once for each item in the sequence.

#### Syntax:
```python
for item in sequence:
    # Code to run for each item
```

#### Iterating Over a List:
Let's see how to print each element in a list.


````python
languages = ["Python", "JavaScript", "SQL", "R"]

print("Iterating over a list:")
for i in languages:
    print(f"I love coding in {i}!")
````

#### The `range()` Function:
In Python, the built-in `range()` function generates a sequence of numbers, which is extremely useful for running a loop a specific number of times.

Syntax: `range(start, stop, step)`
* `start` (optional, default `0`): The beginning number (inclusive).
* `stop` (required): The ending number (exclusive).
* `step` (optional, default `1`): The increment value.

Let's see different ways to use `range()`:


````python
# Generate numbers 0 to 4
print("range(5):", list(range(5)))

# Generate numbers 2 to 8 (exclusive)
print("range(2, 8):", list(range(2, 8)))

# Generate even numbers between 2 and 10
print("range(2, 11, 2):", list(range(2, 11, 2)))

# Generate a countdown from 5 to 1
print("range(5, 0, -1):", list(range(5, 0, -1)))
````

````python
# Using range() in a for loop to calculate the sum of first 5 positive integers (1 to 5)
total = 0
for i in range(10):
    total += i # total=total+1
    # print(f"Added {i} to total, running total: {total}")

print(f"Final Sum: {total}")
````

````python
prime=[]
composite=[]
for i in range(1,21):
    count=0
    for j in range(1,i+1):
        if(i%j==0):
            count+=1
            if(count>2):
                break
    if(count<=2):
        prime.append(i)
    else:
        composite.append(i)    
print(prime)
print(composite)
````

#### Iterating Over a String:
Since a string is a sequence of characters, you can iterate over it letter-by-letter.


````python
word = "Data"
print(f"Letters in '{word}':")
for letter in word:
    print(letter,end=" ")
````

### 2.2 The `while` Loop
A `while` loop runs a block of code as long as a specified condition remains `True`.

#### Syntax:
```python
while condition:
    # Code to run
```

#### ⚠️ Important:
Inside a `while` loop, you must write code that will eventually make the condition `False`. Otherwise, you will create an **infinite loop**, causing your program to run forever!

#### How it works:
```text
           [Start]
              │
              ▼
   ┌──► Is condition True? ── No ──┐
   │          │                    │
   │         Yes                   │
   │          │                    │
   │          ▼                    │
   │    [Run loop-block]           │
   │          │                    │
   └──────────┘                    │
                                   ▼
                              [Exit Loop]
```

Let's see a simple countdown using a `while` loop:


````python
count = 0

while count < 5:
    print(f"{count}")
    count += 1  


````

## Module 3: Loop Control Statements
Sometimes, you need to alter the standard execution of a loop. Python provides three keywords to control loops:
1. **`break`**: Exits the loop prematurely.
2. **`continue`**: Skips the rest of the current iteration and jumps directly to the next iteration.
3. **`pass`**: A null statement used as a placeholder.


### 3.1 The `break` Statement
The `break` statement immediately terminates the loop. Any code after `break` inside the loop is skipped, and the program moves to the statement right after the loop.

#### Example Scenario:
Imagine searching for a specific book in a list. Once you find it, there is no need to search the rest of the list. We use `break` to exit early.


````python
items = ["apple","bread",  "banana", "milk", "chocolate", "eggs"]
search_for = "bread"

print("Starting supermarket search...")
for item in items:
    print("still searching")
    if item == search_for:
        print(f"Found it! '{search_for}' is in the list.")
        break  # Exit the loop immediately
````

### 3.2 The `continue` Statement
The `continue` statement skips the rest of the code in the **current** iteration and moves control back to the top of the loop for the next iteration. It does not stop the entire loop, only the current cycle.

#### Example Scenario:
Print all numbers from 1 to 5, except the number 3.


````python
print("Using continue inside a for loop:")
for num in range(1, 6):
    if num == 3:
        continue  # Skip to the next iteration (num = 4)
    print(f"Number: {num}")
````

````python
# Using continue inside a while loop (Be careful to update the counter BEFORE calling continue!)
num = 0
while num < 5:
    num += 1  # Increment first
    if num == 3:
        print("Skipping 3!")
        continue
    print(f"Value: {num}")
````

### 3.3 The `pass` Statement
The `pass` statement does absolutely nothing! It is a null operation. It is used as a syntax placeholder when a statement is required, but you haven't written the code for it yet.

#### Why do we need it?
Python will raise an error if an empty block is left under an `if` statement or a loop. Using `pass` prevents this syntax error.


````python
# Setting up structure for future code
number = 11

if number > 10:
    pass  # Keeps Python interpreter happy and avoids SyntaxError
else:
    print("Number is 10 or less")
````
