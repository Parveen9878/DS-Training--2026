## 📝 Exercise 1: Strings (`str`)

- **Q1.1: Name Formatter**
  Write a code snippet that takes a messy name input (e.g., `name = "   praneet bawa   "`), trims the leading/trailing spaces, and capitalizes the first letter of each word (Title Case).
  *Expected output:* `"Praneet Bawa"`

- **Q1.2: Vowel Replacer**
  Given the string `sentence = "python is fun"`, replace all occurrences of the letter `"o"` with an asterisk `"*"` and print the result.
  *Expected output:* `"pyth*n is fun"`

- **Q1.3: Check Substring**
  Given `text = "Data Science Plan"`, check if the word `"Science"` exists in the string. Print `True` if it is present, and `False` otherwise.



```python
# Name Formatter

name = "   praneet bawa   "

name = name.strip()
name = name.title()

print("Name:", name)
```

```python
# Vowel Replacer

sentence = "python is fun"

sentence = sentence.replace("o", "*")

print("Modified sentence:", sentence)
```

```python
# Check Substring

text = "Data Science Plan"

print("Science" in text)
```

## 📝 Exercise 2: Lists (`list`)

- **Q2.1: Color Swapper**
  Start with the list `colors = ["red", "green", "blue"]`.
  1. Replace the middle element (`"green"`) with `"yellow"`.
  2. Add `"purple"` to the end of the list.
  3. Print the updated list.
  *Expected output:* `['red', 'yellow', 'blue', 'purple']`

- **Q2.2: List Math**
  Given the list `numbers = [12, 45, 2, 89, 30]`, write code to find and print:
  1. The total count of elements (length).
  2. The maximum value in the list.
  3. The sum of all elements.

- **Q2.3: Slice and Reverse**
  Given `items = ["pen", "pencil", "eraser", "ruler", "marker"]`, extract a slice containing `["pencil", "eraser", "ruler"]` and reverse its elements *without* modifying the original list.

```python
# Color Swapper

colors = ["red", "green", "blue"]

colors[1] = "yellow"

colors.append("purple")

print("Updated colors:", colors)
```

```python
# List Math

numbers = [12, 45, 2, 89, 30]

print("Length of numbers:", len(numbers))

print("Maximum value:", max(numbers))

print("Sum of numbers:", sum(numbers))
```

```python
# Slice and Reverse

items = ["pen", "pencil", "eraser", "ruler", "marker"]

new_list = items[1:4]
new_list.reverse()

print("Reversed slice: ",new_list)

print("Original items: ", items)
```
