### Exercise 4: Simple Grade Calculator (Comparison & Logical)
Given two scores: `homework_score = 85` and `exam_score = 72`.
Write a script that decides if a student passes:
* A student passes if both conditions are met:
  1. `homework_score` is greater than or equal to 80.
  2. `exam_score` is greater than or equal to 70.
* Alternatively, if their `exam_score` is a perfect 100, they pass regardless of their homework score.
Print whether the student passed or failed.

````python
# Exercise 1: Leap Year & Even Check

year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")

if year % 2 == 0:
    print(year, "is an even number.")
else:
    print(year, "is an odd number.")
````

````python
# Exercise 2: Email Domain Checker

allowed_domains = ["gmail.com", "yahoo.com", "outlook.com"]

email = input("Enter your email: ")

if "@" in email:
    domain = email.split("@")[1]

    if domain in allowed_domains:
        print("Allowed email domain.")
    else:
        print("Email domain is not allowed.")
else:
    print("Please enter a valid email address.")
````

````python
# Exercise 3: Value vs Identity Checker

x = [10, 20]
y = [10, 20]

print("x == y :", x == y)
print("x is y :", x is y)

z = x

print("x is z :", x is z)

print("\nExplanation:")
print("x and y have the same values but are different objects.")
print("z refers to the same object as x.")
````

````python
# Exercise 4: Simple Grade Calculator

homework_score = 85
exam_score = 72

if (homework_score >= 80 and exam_score >= 70) or exam_score == 100:
    print("Student Passed")
else:
    print("Student Failed")
````

### Exercise 7: Sum of Even Numbers with early stop
Write a program that:
1. Loops through numbers from 1 to 50.
2. Keeps a running sum of all **even** numbers.
3. If the running sum exceeds **150**, stop the loop using `break` and print the final sum and the number where it stopped.

### Exercise 8: Number filtering using continue
Write a program that:
1. Loops through numbers from 1 to 20.
2. Prints all numbers except numbers that are multiples of 3 or 5 (Hint: use `%` operator and `continue`).


````python
# Exercise 5: Number Classifier

number = float(input("Enter a number: "))

if number > 0:
    print(f"{number} is Positive.")
elif number < 0:
    print(f"{number} is Negative.")
else:
    print(f"{number} is Zero.")
````

````python
# Exercise 6: Multiplication Table Generator

number = int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{number} x {i} = {number * i}")
````

````python
# Exercise 7: Sum of Even Numbers with Early Stop

total = 0

for num in range(1, 51):
    if num % 2 == 0:
        total = total + num

        if total > 150:
            print("Final Sum:", total)
            print("Stopped at Number:", num)
            break
````

````python
# Exercise 8: Number Filtering using continue

for num in range(1, 21):

    if num % 3 == 0 or num % 5 == 0:
        continue

    print(num)
````
